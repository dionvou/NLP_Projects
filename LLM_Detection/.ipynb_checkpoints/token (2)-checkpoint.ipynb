```python
from sklearn.utils import shuffle
df = shuffle(df).reset_index(drop=True)
df.shape
```


Lets now tokenize our essays. We use n-gram (3,5)

```python
# Vectorize
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

tf = TfidfVectorizer(ngram_range=(3, 5),
                             tokenizer=lambda x: re.findall(r'[^\W]+', x),
                             token_pattern=None,
                             lowercase=False,
                             strip_accents='unicode')
vectorizer = tf.fit(df['text'])
X_tfidf = tf.transform(df['text'])
X_tfidf.shape
y = df['generated']
```

```python
# Vectorize test set
test_tfidf = vectorizer.transform(test['text'])
test_y = test['label']
```

Now we will train our model using Multinomial Naive Bayes.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics
from sklearn.metrics import make_scorer, f1_score

# Use f1_score as the scoring metric
scoring_metric = make_scorer(f1_score)

param_grid = {
    'alpha': [0.05, 0.1, 0.2, 0.5, 1.0, 1.5, 2.0],
}

grid_nb = GridSearchCV(MultinomialNB(), param_grid, cv=5, refit=True, verbose=1, scoring=scoring_metric)
grid_nb = grid_nb.fit(X_tfidf, df['generated'])

predicted = grid_nb.best_estimator_.predict(test_tfidf)
print("Classification report for %s:\n%s\n" % (grid_nb.best_estimator_, metrics.classification_report(test_y, predicted)))
```


*   So we can see that the recall of student essays is almost perfect. We need to improve the recall of generated essays, without destroying the overall acurracy.
*   Lets tune our algorithm by changing the threshold of predictions.


```python
# Find the best threshold
from sklearn.metrics import f1_score

f1_scores = []
threshold = np.arange(0,1,0.01)
for t in threshold:
  proba_predictions = grid_nb.best_estimator_.predict_proba(test_tfidf)
  binary_predictions = (proba_predictions[:, 1] > t).astype(int)
  f1_multiclass = f1_score(test_y, binary_predictions, average='weighted')
  f1_scores.append(f1_multiclass)
```

```python
plt.plot(threshold,f1_scores)
plt.axvline(x=threshold[np.argmax(f1_scores)],color='black', linestyle='--', label=f'threshold={threshold[np.argmax(f1_scores)]}')
print(f'Best threshold: {threshold[np.argmax(f1_scores)]}')
plt.xlabel("threshold");
plt.ylabel("f1-score");
plt.title("Best threshold to improve f1 score")
plt.legend()
plt.show()
```

```python
threshold = 0.46
proba_predictions = grid_nb.best_estimator_.predict_proba(test_tfidf)
binary_predictions = (proba_predictions[:, 1] > threshold).astype(int)

print("Classification report for %s:\n%s\n" % (grid_nb.best_estimator_, metrics.classification_report(test_y, binary_predictions)))
```

We made some small improvements:

*   Recall of generated texts from 0.75 to 0.79
*   F1 score from 0.86 to 0.87
*   Accuracy from 0.89 to 0.90







Lets now calculate the similarities of the generated texts with the original texts. We will compute average similarity and max similarity.

```python
df['max_similarity'] = 0
df['avg_similarity'] = 0
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Take indexes where text is generated and not
indices_0 = np.where(df['generated']==0)[0]
indices_1 = np.where(df['generated'] == 1)[0]


for i in tqdm(indices_1, desc="Processing", unit="iteration"):

  # Calculate the similarity of text i with all non-generated texts
  similarity = cosine_similarity(X_tfidf[i],X_tfidf[indices_0])
  df.at[i, 'max_similarity'] = np.max(similarity)
  df.at[i, 'avg_similarity'] = np.mean(similarity)
df.shape
```

```python
from sklearn.model_selection import LeaveOneOut

df['proba'] = 0
loo = LeaveOneOut()

# Cross validation with Leave one out Method
for train_index, test_index in tqdm(loo.split(X_tfidf,df['generated']),total=2842):

    # Split the data into training and test sets
    X_train, X_test = X_tfidf[train_index], X_tfidf[test_index]
    y_train, y_test = df['generated'].iloc[train_index], df['generated'].iloc[test_index]

    # Train the the estimator
    grid_nb.best_estimator_.fit(X_train, y_train)

    # Make predictions
    proba = grid_nb.best_estimator_.predict_proba(X_test)[:,1]

    # Store the prediction to dataframe
    df.at[test_index[0], 'proba'] = proba[0]

```

```python
# Create subplots
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

# Plot data on the first subplot
axs[0].scatter(df['proba'].iloc[indices_1],df['avg_similarity'].iloc[indices_1])
axs[0].set_title('Correlation of average similarity with probability \nto be generated ')
axs[0].set_ylabel('Average Similarity')
axs[0].set_xlabel('Probability')

# Plot data on the second subplot
axs[1].scatter(df['proba'].iloc[indices_1],df['max_similarity'].iloc[indices_1])
axs[1].set_title('Correlation of maximum similarity with probability \nto be generated ')
axs[1].set_ylabel('Max Similarity')
axs[1].set_xlabel('Probability')


plt.tight_layout()
plt.show()
```

```python
print(f'Correlation between predict probability and max_similarity: {df["proba"].iloc[indices_1].corr(df["max_similarity"].iloc[indices_1])}')
```

```python
# For small dataset
from sklearn.metrics import accuracy_score

training_sizes = np.arange(0.05, 1.05, 0.05)

classifier = MultinomialNB(alpha=0.5)

train_accuracies, test_accuracies = [], []
for size in training_sizes:
  subset_size = int(size * X_tfidf.shape[0])
  X_subset, y_subset = X_tfidf[:subset_size], y[:subset_size]

  classifier.fit(X_subset, y_subset)

  y_train_pred = classifier.predict(X_tfidf)
  train_acc = accuracy_score(y, y_train_pred)
  train_accuracies.append(train_acc)

  y_test_pred = classifier.predict(test_tfidf)
  test_acc = accuracy_score(test_y, y_test_pred)
  test_accuracies.append(test_acc)
```

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(training_sizes.reshape(-1, 1), test_accuracies)

# Step 6: Estimate Number of Texts for "Best" Performance
desired_performance = 0.97
estimated_size = (desired_performance - regressor.intercept_) / regressor.coef_[0]
new_training = np.arange(0.1, 1.9, 0.1)
predictions = regressor.predict(new_training.reshape(-1, 1))
estimated_texts_needed = (estimated_size * X_tfidf.shape[0]) -X_tfidf.shape[0]
print(f'Estimated more texts needed for accuracy 97% : {estimated_texts_needed:.0f}')
```

```python
# Plot learning curves
plt.plot(new_training,predictions,label='Estimation accuracy rate')
plt.plot(training_sizes, train_accuracies, label='Train Accuracy')
plt.plot(training_sizes, test_accuracies, label='Test Accuracy')
plt.xlabel('Training Size')
plt.ylabel('Accuracy')
plt.title('Learning Curves')
plt.legend()
plt.xlim(0,2)
plt.axvline(x=estimated_size, color='b', linestyle='--',lw='1')

plt.show()
```

Seems that improves fast with more data.

```python
data_0 = data[data.generated==0]
data_1 = data[data.generated==1]
```

```python
X_tfidf_0 = tf.transform(data_0['text'])
X_tfidf_1 = tf.transform(data_1['text'])
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

silhouette_scores_0 = []
K = range(2, 5)
for k in tqdm(K):
    kmeanModel = KMeans(n_clusters=k,n_init=8)
    kmeanModel.fit(X_tfidf_0)
    score = silhouette_score(X_tfidf_0, kmeanModel.labels_)
    silhouette_scores_0.append(score)

silhouette_scores_1 = []
K = range(2, 5)
for k in tqdm(K):
    kmeanModel = KMeans(n_clusters=k,n_init=8)
    kmeanModel.fit(X_tfidf_1)
    score = silhouette_score(X_tfidf_1, kmeanModel.labels_)
    silhouette_scores_1.append(score)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

# Subplot 1
ax1.bar(K, silhouette_scores_0)
ax1.set_xlabel('Number of clusters')
ax1.set_ylabel('Silhouette Score')
ax1.set_title('Silhouette Score for Student essays')
ax1.set_xticks(K)


# Subplot 2
ax2.bar(K, silhouette_scores_1)
ax2.set_xlabel('Number of clusters')
ax2.set_ylabel('Silhouette Score')
ax2.set_title('Silhouette Score for generated essays')
ax2.set_xticks(K)


plt.tight_layout()
plt.show()
```

We will keep K = 2, which is also the number of prompts.

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2)
# Kmeans for generated = 0
kmeans.fit(X_tfidf_0)
clusters_0 = kmeans.predict(X_tfidf_0)

# Kmeans for generated = 1
kmeans.fit(X_tfidf_1)
clusters_1 = kmeans.predict(X_tfidf_1)
```

```python
from sklearn.decomposition import TruncatedSVD

# Generated = 0 SVD_with 2 components
svd = TruncatedSVD(n_components=2)
svd_result_0 = svd.fit_transform(X_tfidf_0)

# Create a DataFrame with the results and cluster labels
df_svd_0 = pd.DataFrame(data=svd_result_0, columns=['SVD1', 'SVD2'])
df_svd_0['Cluster'] = clusters_0

# Generated = 0 SVD_with 2 components
svd = TruncatedSVD(n_components=2)
svd_result_1 = svd.fit_transform(X_tfidf_1)

# Create a DataFrame with the results and cluster labels
df_svd_1 = pd.DataFrame(data=svd_result_1, columns=['SVD1', 'SVD2'])
df_svd_1['Cluster'] = clusters_1
```

```python
from matplotlib.lines import Line2D


# Visualize the clusters in 2D
fig, axs = plt.subplots(1, 2, figsize=(15, 6))

# Plot for the first subplot
scatter1 = axs[0].scatter(df_svd_0['SVD1'], df_svd_0['SVD2'], c=df_svd_0['Cluster'], cmap='viridis')
axs[0].set_xlabel('SVD1')
axs[0].set_ylabel('SVD2')
axs[0].set_title('Subplot 1')

# Plot for the second subplot
scatter2 = axs[1].scatter(df_svd_1['SVD1'], df_svd_1['SVD2'], c=df_svd_1['Cluster'], cmap='viridis')
axs[1].set_xlabel('SVD1')
axs[1].set_ylabel('SVD2')
axs[1].set_title('Subplot 2')

# Customize the legend labels
new_labels_0 = [f"Prompt:0, count:{df_svd_0['Cluster'].value_counts()[0]}", f"Prompt:1, count:{df_svd_0['Cluster'].value_counts()[1]}"]
new_labels_1 = [f"Prompt:0, count:{df_svd_1['Cluster'].value_counts()[0]}", f"Prompt:1, count:{df_svd_1['Cluster'].value_counts()[1]}"]
# Add legend for the first subplot
legend_elements1 = [Line2D([0], [0], marker='o', color='w', markerfacecolor=plt.cm.viridis(i / max(df_svd_0['Cluster'])), markersize=5, label=new_labels_0[i]) for i in range(2)]
axs[0].legend(handles=legend_elements1, title='Clusters')

# Add legend for the second subplot
legend_elements2 = [Line2D([0], [0], marker='o', color='w', markerfacecolor=plt.cm.viridis(i / max(df_svd_0['Cluster'])), markersize=5, label=new_labels_1[i]) for i in range(2)]
axs[1].legend(handles=legend_elements2, title='Clusters')

plt.suptitle('KMeans Clustering with 2 SVD Components')
plt.show()
```
